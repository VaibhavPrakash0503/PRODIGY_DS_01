## Importing libraries and the dataset

```python
import pandas as pd

#Reading the data set
df = pd.read_csv("data.csv")

#Removing the unnecessary values
df = df.drop(columns=["Indicator Name", "Indicator Code", "Country Name"])
```

# Removing the missing values

```python
df = df.fillna(0)
```

## Countries Based on total population

```python
sorted_population = df.sort_values(by='2023',ascending=False)

# Removing the population count of continent and the sub-continent
sorted_population=sorted_population[15:]

# print("Top 10 countires of highest population")
top_10_count = sorted_population.head(10)
print(top_10_count["Country Code"])
```

## Bar Chart of Top 5 countries

```python
import matplotlib.pyplot as plt

#Top ten Countries for year 2023
top_10_count.plot(x="Country Code", y = "2023", kind = "bar", figsize=(8,5),color = "skyblue")
plt.title("Population Trends of top 10 Countries")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.show()
```

## Line Chart of Population of countries over the year

```python
top_10_count.loc[["IND","CHN"]].T.plot(kind="line",figsize=(10,6),color=["orange","red","blue"])

plt.title("Population trend in India and China over the years")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.legend(title="Country")
plt.show()
```

## Area Plot of population of India

```python
top_10_count.loc["IND"].plot(kind="area",figsize=(10,6),color="skyblue")

plt.title("Population of India over the years")
plt.xlabel("Year")
plt.ylabel("Poupulation")
plt.show()
```

## Population Comparison in year 2023

```python
country = top_10_count.loc[["IND","CHN"],"2023"]

country.plot(kind="bar",figsize=(10,6), color=["orange","red"],width=0.4)

plt.title("Population comparison in year 2023")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```
